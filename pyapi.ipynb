```python
# Create and write to a .py file
code = """
from fastapi import FastAPI
app = FastAPI()

@app.get("/")
def read_root():
  return {"message": "Hello from FastAPI in Colab!"}

@app.get("/hello/{name}")
def hello(name: str):
  return {"message": f"Hello, {name}"}
"""

with open("my_script.py", "w") as f:
    f.write(code)
```
